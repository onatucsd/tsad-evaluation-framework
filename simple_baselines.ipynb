```python
sensor_baseline = simple_baselines.SensorRangeDeviation()
norm_baseline = simple_baselines.LNorm() 
neighbor_baseline = simple_baselines.NNDistance()
pca_baseline = simple_baselines.PCAError()
```

```python
sensor_baseline.fit(SMD.train)
sensor_preds = sensor_baseline.transform(SMD.test)
dyn_th_calculate_metrics(SMD.test_labels, sensor_preds)
```

```python
norm_baseline.fit(SMD.train)
norm_scores = norm_baseline.transform(SMD.test)
opt_th = best_f_score(norm_scores, SMD.test_labels)
calculate_metrics(SMD.test_labels, norm_scores, opt_th)
```

```python
pca_baseline.fit(SMD.train)
pca_errors = pca_baseline.transform(SMD.test)
pca_scores = np.mean(pca_errors, axis=1) # MoC
opt_th = best_f_score(pca_scores, SMD.test_labels)
calculate_metrics(SMD.test_labels, pca_scores, opt_th)
```
